## Linear Regression for Diabetes dataset - Lesson 1

Import needed libraries

```python
# Import needed libraries
import matplotlib.pyplot as plt  # Import the matplotlib library for plotting graphs.
import numpy as np  # Import the numpy library for numerical operations.
from sklearn import datasets, linear_model, model_selection  # Import specific modules from scikit-learn for datasets, linear regression model, and model selection utilities.

```

Load the diabetes dataset, divided into `X` data and `y` features

```python
# Load the diabetes dataset, divided into X data and y features
X, y = datasets.load_diabetes(return_X_y=True)  # Load the diabetes dataset, returning the feature matrix X and target vector y.
# Print the shape of X and first element of y
print(X.shape)  # Print the shape of X, which is (442, 10), indicating 442 samples and 10 features.
print(y[0])  # Print the first element of y to see an example target value.
```

Select just one feature to target for this exercise

```python
# Select one feature to target for this exercise
X = X[:, np.newaxis, 2]  # Select the third feature (index 2) from the dataset and add a new axis to make X a 2D array with shape (442, 1).

```

Split the training and test data for both `X` and `y`

```python
# Split the training and test data for both X and y
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)  # Split the data into training and testing sets, with 33% of the data used for testing.
```

Select the model and fit it with the training data

```python
# Select the model and fit it with the training data
model = linear_model.LinearRegression()  # Create an instance of the LinearRegression model.
model.fit(X_train, y_train)  # Fit the model using the training data.
```

Use test data to predict a line

```python
# Use test data to predict a line
y_pred = model.predict(X_test)  # Predict the target values for the test set using the fitted model.

```

Display the results in a plot

```python
# Display the results in a plot
plt.scatter(X_test, y_test, color='black')  # Plot the test data points as black dots.
plt.plot(X_test, y_pred, color='blue', linewidth=3)  # Plot the regression line (predicted values) in blue.
plt.show()  # Display the plot.
```
